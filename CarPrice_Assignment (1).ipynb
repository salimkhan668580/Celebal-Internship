```python
name = df["Brand"].head(20)
price = df["Selling_Price"].head(20)

fig, ax = plt.subplots(figsize =(16, 9))

# Horizontal Bar Plot
ax.barh(name, price)

# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

# Show top values
ax.invert_yaxis()

# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')

# Add Plot Title
ax.set_title('Sports car and their price',
             loc ='center')

plt.show()
```

```python
name = df["Brand"].head(20)
Selling_Price = df["Selling_Price"].head(20)

df_subset = pd.DataFrame({'Brand':name, 'Selling_Price': Selling_Price})

df_subset.plot(x='Brand', y='Selling_Price', kind='barh', stacked=True)
plt.title('Cars and their Horse Power')
plt.xlabel('Selling_Price', size=15)
plt.ylabel('Brand', size=15)
plt.show()
```

```python
name = df["Brand"].head(20)
KM_Driven = df["KM_Driven"].head(20)

# Defining the plotsize
plt.figure(figsize=(8, 6))

df_subset = pd.DataFrame({'Brand':name, 'KM_Driven':KM_Driven})

# Defining the x-axis, the y-axis and the data
# from where the values are to be taken
plots = sns.barplot(x="KM_Driven", y="Brand", data=df_subset)

# Setting the x-acis label and its size
plt.xlabel("Curb Weight", size=15)

# Setting the y-axis label and its size
plt.ylabel("Brand", size=15)

plt.title("CarName and their Curb Weight")

# Finallt plotting the graph
plt.show()
```

```python
price = df["Selling_Price"].head(20)

data_point = df["Selling_Price"].tolist()

fig, ax = plt.subplots()
ax.hist(data_point, bins=20)

plt.show()
```

```python
price = df["Selling_Price"].head(20)

data_point = df["Selling_Price"].tolist()

np.random.seed(42)
data_point = np.random.normal(loc=50, scale=10, size=1000)

# Customizing the histogram
fig, ax = plt.subplots(figsize=(8, 6))
ax.hist(data_point, bins=20, color='skyblue', edgecolor='black', alpha=0.7)

# Adding labels and title
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Customized Histogram', fontsize=14)

# Modifying axis ticks
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Adding gridlines
plt.grid(True, linestyle='--', alpha=0.5)

# Finally, displaying the plot
plt.show()
```

```python
name = df["Brand"].head(20)
price = df["Selling_Price"].head(20)

# Calculate the total price for each car name
total_price_by_name = price.groupby(name).sum()

# Create pie plot
fig, ax = plt.subplots()
ax.pie(total_price_by_name, labels=total_price_by_name.index, autopct='%1.1f%%',
        wedgeprops = {"edgecolor" : "black", #here we are providing borders around
                      'linewidth': 1,
                      'antialiased': True})

# Set aspect ratio to be equal, creating a circular pie plot
ax.set_aspect('equal')

# Set plot title
ax.set_title("Distribution of Car Prices by Name")

# Show the plot
plt.show()
```

```python
Brand = df["KM_Driven"].head(100)
Selling_Price = df["Selling_Price"].head(100)
Year = df["Year"].head(100)

# Create 3D scatter plot
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

ax.plot3D(Brand,Selling_Price,Year, "red")

# Set axis labels
ax.set_xlabel("KM_Driven")
ax.set_ylabel("Price")
ax.set_zlabel("Year")

# Set plot title
ax.set_title("3D Dimensional Plot")

# Adjust plot boundaries
ax.auto_scale_xyz(Brand, Selling_Price,Year)

# Show the plot
plt.show()
```
